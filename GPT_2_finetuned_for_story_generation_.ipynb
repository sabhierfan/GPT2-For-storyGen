```python
import torch
from transformers import GPT2Tokenizer, GPT2LMHeadModel, Trainer, TrainingArguments
from datasets import load_dataset

# Load model and tokenizer
model_name = "gpt2"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

ds = load_dataset("raviraaj/text_to_image_prompt")

# Preprocess the dataset
def tokenize_function(examples):
    # The key for the text data is likely 'Text' not 'Captions'
    return tokenizer(examples["Text"], padding="max_length", truncation=True, max_length=128)

# Tokenize the dataset
tokenized_dataset = ds.map(tokenize_function, batched=True)
tokenized_dataset.set_format(type="torch", columns=["input_ids", "attention_mask"])

# Define training arguments
training_args = TrainingArguments(
    output_dir="./fine_tuned_model",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=4,
    num_train_epochs=3,
    weight_decay=0.01,
)

# Create Trainer instance
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
)

# Fine-tune the model
trainer.train()

# Save the fine-tuned model
trainer.save_model("fine_tuned_prompt_enhancer")
tokenizer.save_pretrained("fine_tuned_prompt_enhancer")
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("describeai/gemini")
model = AutoModelForSeq2SeqLM.from_pretrained("describeai/gemini")
```

```python
# Set the pad_token to eos_token
tokenizer.pad_token = tokenizer.eos_token
```

```python
def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True, return_tensors="pt")
```

```python
# Apply the tokenizer to the dataset
tokenized_datasets = ds.map(tokenize_function, batched=True)
```

```python
# Function to format the dataset correctly
def format_dataset(dataset):
    dataset = dataset.map(lambda examples: {
        'input_ids': examples['input_ids'],
        'attention_mask': examples['attention_mask'],
        'labels': examples['input_ids']
    }, batched=True)
    return dataset
```

```python
tokenized_datasets = format_dataset(tokenized_datasets)
```

```python
# Split the dataset manually into train and validation sets
train_size = int(0.8 * len(tokenized_datasets['train']))
test_size = len(tokenized_datasets['train']) - train_size

train_dataset = tokenized_datasets['train'].select(range(train_size))
eval_dataset = tokenized_datasets['train'].select(range(train_size, len(tokenized_datasets['train'])))
```

```python
# Load a pre-trained model
model = AutoModelForSeq2SeqLM.from_pretrained("describeai/gemini")

# Set up training arguments
training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    learning_rate=5e-5,
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
    num_train_epochs=1,
    weight_decay=0.01,
)
```

```python
# Initialize the Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
)

# Train the model
trainer.train()
```

```python
# Load the fine-tuned model for text generation
story_generator = pipeline('text-generation', model=model, tokenizer=tokenizer)
```

```python
# Generate a story based on a prompt
prompt = "Once upon a time in a magical forest, there was a young unicorn named Lily."
generated_story = story_generator(prompt, max_length=200, num_return_sequences=1)
```
